```python
def player(prev_play, opponent_history=[]):

    if prev_play:
        opponent_history.append(prev_play)


    last_opponent_play = prev_play


    if last_opponent_play == 'R':
        return 'P'
    elif last_opponent_play == 'P':
        return 'S'
    elif last_opponent_play == 'S':
        return 'R'
    else:

        import random
        return random.choice(['R', 'P', 'S'])


```
